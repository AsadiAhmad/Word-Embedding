# Step5: Create Model for Vectorization (word Embeding)

## Step5.1: Embedding Model

```python
def get_embedding_model(train_data, test_data, train_labels, test_labels, max_words=25000, embedding_dim=64, input_length=50, num_clusters=2):
    train_padded = pad_sequences(train_data, maxlen=input_length)
    test_padded = pad_sequences(test_data, maxlen=input_length)

    model = Sequential()
    model.add(Embedding(max_words, embedding_dim, input_length=input_length))
    model.add(Flatten())
    model.add(Dense(embedding_dim, activation='linear'))

    model.compile(optimizer='adam', loss='mse')

    train_embeddings = model.predict(train_padded)

    test_embeddings = model.predict(test_padded)

    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    clusters = kmeans.fit_predict(test_embeddings)

    binary_predictions = to_categorical(clusters)

    return binary_predictions
```

```python
def word_to_vec(train_data, test_data, train_labels):
    word2vec_model = Word2Vec.load("/content/word2vec_model.bin")

    train_vectors = []
    for sentence in train_data:
        vectors = [word2vec_model.wv[word] if word in word2vec_model.wv else np.zeros(word2vec_model.vector_size) for word in sentence]
        train_vectors.append(np.mean(vectors, axis=0))

    test_vectors = []
    for sentence in test_data:
        vectors = [word2vec_model.wv[word] if word in word2vec_model.wv else np.zeros(word2vec_model.vector_size) for word in sentence]
        test_vectors.append(np.mean(vectors, axis=0))

    train_vectors = np.array(train_vectors)
    test_vectors = np.array(test_vectors)
    train_labels = np.array(train_labels)

    knn_classifier = KNeighborsClassifier(n_neighbors=5)
    knn_classifier.fit(train_vectors, train_labels)

    binary_predictions = knn_classifier.predict(test_vectors)

    return binary_predictions
```

## Step5.3: Create SimpleRNN Model

```python
def get_SimpleRNN_model(train_data, test_data, train_labels, test_labels, max_words=25000, embedding_dim=64, input_length=50, num_clusters=2):
    train_padded = pad_sequences(train_data, maxlen=input_length)
    test_padded = pad_sequences(test_data, maxlen=input_length)

    model = Sequential()
    model.add(Embedding(max_words, embedding_dim, input_length=input_length))
    model.add(SimpleRNN(32, return_sequences=True))
    model.add(SimpleRNN(16))
    model.add(Flatten())
    model.add(Dense(embedding_dim, activation='linear'))

    model.compile(optimizer='adam', loss='mse')

    train_embeddings = model.predict(train_padded)

    test_embeddings = model.predict(test_padded)

    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    clusters = kmeans.fit_predict(test_embeddings)

    binary_predictions = to_categorical(clusters)

    return binary_predictions
```

## Step5.4: Create LSTM Model

```python
def LSTM2(train_data, test_data, train_labels, test_labels, max_words=25000, embedding_dim=300, input_length=50, num_clusters=2):
    train_padded = pad_sequences(train_data, maxlen=input_length)
    test_padded = pad_sequences(test_data, maxlen=input_length)

    word2vec_model = Word2Vec.load("/content/word2vec_model.bin")

    embedding_matrix = np.zeros((max_words, embedding_dim))
    for word, i in word_index.items():
        if i < max_words and word in word2vec_model.wv:
            embedding_matrix[i] = word2vec_model.wv[word]

    model = Sequential()
    model.add(Embedding(max_words, embedding_dim, input_length=input_length))
    model.add(LSTM(32, return_sequences=True))
    model.add(LSTM(16))

    model.layers[0].set_weights([embedding_matrix])
    model.layers[0].trainable = False

    model.add(Dense(num_clusters, activation='softmax'))

    model.compile(optimizer='adam', loss='categorical_crossentropy')

    train_embeddings = model.predict(train_padded)
    test_embeddings = model.predict(test_padded)

    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    clusters = kmeans.fit_predict(test_embeddings)

    binary_predictions = to_categorical(clusters)

    return binary_predictions
```

## Step6.2: Add Predicted Column

## Step6.3: Calculate the result of Confusion Matrix

```python
def compute_confusion_matrix_results(confusion_matrix):
    conditions = [
        (confusion_matrix['Label'] == 1) & (confusion_matrix['Predicted'] == 1),  # True Positive (TP)
        (confusion_matrix['Label'] == 0) & (confusion_matrix['Predicted'] == 0),  # True Negative (TN)
        (confusion_matrix['Label'] == 0) & (confusion_matrix['Predicted'] == 1),  # False Positive (FP)
        (confusion_matrix['Label'] == 1) & (confusion_matrix['Predicted'] == 0)   # False Negative (FN)
    ]

    choices = ['TP', 'TN', 'FP', 'FN']

    confusion_matrix['result'] = np.select(conditions, choices, default='')

    return confusion_matrix
```

```python
confusion_matrix11 = compute_confusion_matrix_results(confusion_matrix11)
confusion_matrix12 = compute_confusion_matrix_results(confusion_matrix12)
confusion_matrix13 = compute_confusion_matrix_results(confusion_matrix13)
confusion_matrix14 = compute_confusion_matrix_results(confusion_matrix14)

confusion_matrix21 = compute_confusion_matrix_results(confusion_matrix21)
confusion_matrix22 = compute_confusion_matrix_results(confusion_matrix22)
confusion_matrix23 = compute_confusion_matrix_results(confusion_matrix23)
confusion_matrix24 = compute_confusion_matrix_results(confusion_matrix24)
```

# Step7: Calculate Measurments

```python
def calculate_metrics(confusion_matrix):
    TP = confusion_matrix[confusion_matrix['result'] == 'TP'].shape[0]
    FP = confusion_matrix[confusion_matrix['result'] == 'FP'].shape[0]
    TN = confusion_matrix[confusion_matrix['result'] == 'TN'].shape[0]
    FN = confusion_matrix[confusion_matrix['result'] == 'FN'].shape[0]

    precision = TP / (TP + FP) if TP + FP != 0 else 0
    recall = TP / (TP + FN) if TP + FN != 0 else 0
    accuracy = (TP + TN) / (TP + TN + FP + FN) if TP + TN + FP + FN != 0 else 0

    return precision, recall, accuracy
```

```python
precision11, recall11, accuracy11 = calculate_metrics(confusion_matrix11)
precision12, recall12, accuracy12 = calculate_metrics(confusion_matrix12)
precision13, recall13, accuracy13 = calculate_metrics(confusion_matrix13)
precision14, recall14, accuracy14 = calculate_metrics(confusion_matrix14)

precision21, recall21, accuracy21 = calculate_metrics(confusion_matrix21)
precision22, recall22, accuracy22 = calculate_metrics(confusion_matrix22)
precision23, recall23, accuracy23 = calculate_metrics(confusion_matrix23)
precision24, recall24, accuracy24 = calculate_metrics(confusion_matrix24)
```
